# **Exploratory Data Analysis**

```python
#  Find accident severity by weather condition
plt.figure(figsize=(10, 5))
# Define a custom color palette with shades of blue
blue_palette = sns.color_palette(["#4682b4", "#0000ff", "#5f9ea0", "#00008B"])

sns.countplot(x='Weather_Condition', hue='Severity', data=df, palette=blue_palette)
plt.xlabel('Wether_condition')
plt.ylabel('Count')
plt.title('Accident Severity Distribution by Weather Condition')

plt.show()
```

```python
# Distribution of vehicle type in accidents
colors = ['#0000FF', '#5f9ea0', '#00BFFF', '#87CEEB', '#ADD8E6', '#1E90FF']

# Get vehicle type counts
vehicle_counts = df['Vehicle_Type'].value_counts()

# Plot the pie chart with custom blue shades
vehicle_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    title='Distribution of Vehicle Types in Accidents',
    colors=colors,
    startangle=90,  # Start angle for better alignment
    explode=[0.05] * len(vehicle_counts)  # Slightly separate all slices
)
plt.ylabel('')
plt.show()
```

```python
# Accident rate by day of week and hour of day
plt.figure(figsize=(10, 6))
# Pivot table for heatmap
accident_pivot = df.pivot_table(index='Day_of_Week', columns='Hour_of_Day', aggfunc='size')

sns.heatmap(accident_pivot, cmap='Blues', annot=True)
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
plt.title('Heatmap of Accident Rates by Day and Hour')

plt.show()
```

```python
# No of vehicles involved in severe accidents
plt.figure(figsize=(10, 6))
blue_palette = sns.color_palette(["#6495ED", "#191970", "#1E90FF", "#007BA7"])
sns.countplot(x='Vehicle_Type', hue='Severity', data=df, palette=blue_palette)
plt.title('Vehicle Types Involved in Severe Accidents', fontsize=14)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Accidents frequency by season and day of the week
plt.figure(figsize=(9, 4))

# Define a color palette with enough shades of blue for each day of the week
blue_palette = sns.color_palette(["#6495ED", "#191970", "#1E90FF", "#007BA7"])

# Create a FacetGrid
g = sns.FacetGrid(df, col='Season', height=5, aspect=1.2)

# Map the countplot with enough colors in the palette
g.map(sns.countplot, 'Day_of_Week', order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
      color=blue_palette[0])  # Use a consistent color if `hue` is unnecessary

# Add titles and labels
g.set_titles('{col_name}')
g.set_axis_labels('Day_of_Week', 'Count')

# Add a common title
g.fig.suptitle('Accident Frequency by Season and Day of the Week', fontsize=18)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
# No of accidents by hour of the day and season
aggregated_data = df.groupby(['Hour_of_Day', 'Season']).size().reset_index(name='Accident_Count')

# Plot the line chart
plt.figure(figsize=(10, 5))
sns.lineplot(
    x='Hour_of_Day',
    y='Accident_Count',
    hue='Season',
    data=aggregated_data,
    marker='o',
    palette='cool'
)
# Customize the chart
plt.title('Accidents by Hour of the Day and Season', fontsize=14)
plt.xlabel('Hour of the Day', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)
plt.xticks(range(0, 24))  # Ensure all hours are shown
plt.legend(title='Season', title_fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Find the accident severity by no of vehicles involved
plt.figure(figsize=(12, 6))

# Create a violin plot for accident severity by the number of vehicles involved, with hue for seasons
sns.violinplot(
    x='Number_of_Vehicles',
    y='Severity',
    hue='Season',  # Use 'Season' as the hue variable
    data=df,
    palette='Blues',
    split=True  # Splits the violin for each hue
)

# Add title and axis labels
plt.title('Accident Severity by Number of Vehicles Involved (Grouped by Season)', fontsize=14)
plt.xlabel('Number of Vehicles Involved', fontsize=12)
plt.ylabel('Severity', fontsize=12)

# Adjust legend position for better clarity
plt.legend(title='Season', loc='upper right', bbox_to_anchor=(1.15, 1))

# Display the plot
plt.show()
```

```python
# Impact of weather conditions on number of vehicles involved
plt.figure(figsize=(10, 5))

# Create the scatter plot with blue shades for severity
sns.scatterplot(
    x='Weather_Condition',
    y='Number_of_Vehicles',
    data=df,
    hue='Severity',         # Represents accident severity with color
    palette='Blues',        # Color palette in blue shades
    size='Severity',        # Represents severity with point size
    sizes=(50, 200)         # Range of point sizes
)

# Add title and labels
plt.title('Impact of Weather Conditions on Number of Vehicles Involved', fontsize=14)
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Number of Vehicles', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# Top 10 location with most accidents
plt.figure(figsize=(10, 5))

# Get the top 10 locations with the most accidents
top_locations = df['Location'].value_counts().head(10)

# Define a custom list of 10 shades of blue
blue_palette = ["#00224d", "#003366", "#99c2ff", "#66a3ff", "#3385ff",
    "#0066ff", "#0052cc", "#004099", "#cce0ff", "#e6f2ff"]

# Create the barplot with top locations and accident counts
sns.barplot(x=top_locations.index, y=top_locations.values, color=blue_palette[0])

# Set plot title and axis labels
plt.title('Top 10 Locations with Most Accidents', fontsize=14)
plt.xlabel('Location', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
#Find the severity of accidents by location
heatmap_data = df.pivot_table(index='Location', columns='Severity', aggfunc='size', fill_value=0)

plt.figure(figsize=(10, 5))
sns.heatmap(heatmap_data, annot=True, cmap='Blues', fmt='d')
plt.title('Severity of Accidents by Location', fontsize=14)
plt.xlabel('Severity', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.show()
```

```python
#Find the most common causes of accidents by severity
plt.figure(figsize=(10, 5))

# Create countplot for accidents by cause and severity, with 'Cause' as the correct column name
sns.countplot(x='Cause', data=df, hue='Severity', palette='Blues')

# Add title and labels
plt.title('Accidents by Cause and Severity', fontsize=14)
plt.xlabel('Cause of Accident', fontsize=12)
plt.ylabel('Number of Accidents', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```
